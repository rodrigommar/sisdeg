```python
import requests
import pandas as pd
```

# Eu estou usando uma api do IBGE e coletando dados populacionais de 2022 pra efeito de prototipagem.


```python
url_ibge = 'https://servicodados.ibge.gov.br/api/v3/agregados/4709/periodos/2022/variaveis/93?localidades=N3[all]'

response = requests.get(url_ibge)

if response.status_code == 200:
    dados_populacao = response.json()
else:
    print(f'Error ao acessar: {response.status_code}')

dados_estado = dados_populacao[0]['resultados']
data_list = []

for resultados in dados_estado:
    for serie in resultados['series']:
        localidade = serie['localidade']['nome']
        populacao = serie['serie']['2022'] # para efeito de prototipagem estou usando uma base de 2022
        data_list.append({"Estado": localidade, "População": populacao})

df_populacao = pd.DataFrame(data_list)

```

```python
df_populacao.sort_values('Estado', inplace=True)
```

```python
df_populacao.info()
```

```python
# mudar o tipo de variavel
df_populacao['População'] = df_populacao['População'].astype('Int64')
```

```python
# adicionar as siglas dos estdos
df_populacao['UF'] = ["AC", "AL", "AP", "AM", "BA", "CE", "DF", "ES", "GO", "MA", "MT", "MS", "MG", "PR", "PB", "PA", "PE", "PI", "RN", "RS", "RJ", "RO", "RR", "SC", "SE", "SP", "TO"]
```
