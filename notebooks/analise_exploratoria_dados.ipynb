```python
filter_sp = df[['SG_UF_NOT','DISPNEIA','OBESIDADE',	'DIABETES',	'PNEUMOPATI','IMUNODEPRE']]['SG_UF_NOT'] == 'SP'

filter_com = df.loc[filter_sp, ['DISPNEIA','OBESIDADE',	'DIABETES',	'PNEUMOPATI','IMUNODEPRE']]

for comorbidade in filter_com.columns:

    print(filter_com[comorbidade].value_counts())


```

```python

def grupo_de_riscos(dataframe):

    lista_riscos = ['DISPNEIA','OBESIDADE',	'DIABETES',	'PNEUMOPATI','IMUNODEPRE']
    
    dataframe['COMORBIDADE'] = dataframe[lista_riscos].isin([1]).any(axis=1)
    total_por_estado = dataframe[dataframe['COMORBIDADE']].groupby(
        'SG_UF_NOT').size().reset_index(name='COMORBIDADE')

    return total_por_estado

df_grupo_riscos = grupo_de_riscos(df)
df_grupo_riscos
```
