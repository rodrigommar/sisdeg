```python
from pathlib import Path
import pandas as pd
```

```python
def ler_arquivo_csv(nome_arquivo, sep=',', encoding='latin-1'):

    caminho_notebook = Path.cwd()
    caminho_relativo_csv = f'./data/{nome_arquivo}.csv'
    caminho_raiz = caminho_notebook.parent
    caminho_abs_csv = caminho_raiz / caminho_relativo_csv

    try:
        df = pd.read_csv(caminho_abs_csv, sep=sep, encoding=encoding, low_memory=False)
        return df
    except FileNotFoundError:
        print(f"O arquivo CSV n√£o foi encontrado em {caminho_abs_csv}")
```

#### Criei uma fun√ß√£o **substituindo_valores_nan()** para substituir os valores Nan das colunas float e que tem 3 tipos de valores (1,2,9)

1-Sim
2-N√£o
9-Ignorado

Aqueles valores 'Nan' substitui pelo valor 9.0 (ignorado)

```python
lista_colunas_com_nan = [
    'TOSSE',
    'GARGANTA',
    'DISPNEIA',
    'OBESIDADE',
    'DIABETES',
    'PNEUMOPATI',
    'IMUNODEPRE'
]

def substituindo_valores_nan(dataframe, lista_colunas):

    for coluna in dataframe[lista_colunas]:

        valores = len(list(dataframe[coluna].unique()))

        if (dataframe[coluna].dtypes == 'float64') & (valores == 4):

            filter = dataframe[coluna].isna()
            dataframe.loc[filter, coluna] = dataframe.loc[filter, coluna].fillna(9.0)

    return dataframe

df_versao2 = substituindo_valores_nan(df_versao1, lista_colunas_com_nan)

df_versao2.isna().sum()
```
